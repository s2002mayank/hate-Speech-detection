```python
import numpy as np
import pandas as pd
```

```python
dataset=pd.read_csv("twitter.csv")
dataset
```

```python
dataset.isnull().sum()
```

```python
dataset['labels']= dataset['class'].map({0: "hate speech or offensive language",
                                         1: "hate speech or offensive language",
                                         2: "no hate or offensive language"})
```

```python
data= dataset[["tweet", "labels"]]
```

```python
import re 
import nltk
import string
```

```python
stopwords = set(nltk.corpus.stopwords.words("english"))
```

```python
stopwords.add("rt")
stopwords
```

```python
stemmer= nltk.SnowballStemmer("english")
```

```python
def data_clean(text):
    text=str(text).lower()
    # text=re.sub("https?://S+www\.S+", "", text)
    # text=re.sub("\[.*?\]","", text)
    # text=re.sub("<.*?>+","", text)
    # text=re.sub("[%s]"% re.escape(string.punctuation),"", text)
    text=re.sub("\n","", text)
    # text=re.sub("\w*\d\w*","", text)
    words= [stemmer.stem(word) for word in text.split(' ') if word and word not in stopwords]
    text=" ".join(words)
    return text
    
```

```python
data["tweet"]=data["tweet"].apply(data_clean)
```

```python
X=np.array(data["tweet"])
y=np.array(data["labels"])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
```

```python
cv=CountVectorizer()
X=cv.fit_transform(X)
```

```python
X
```

```python
x_train, x_test, y_train, y_test=train_test_split(X,y, test_size=0.3)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier()
dt.fit(x_train, y_train)
```

```python
y_pred=dt.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
cm=confusion_matrix(y_test,  y_pred)
cm
```

```python
print(classification_report(y_test, y_pred))
```
